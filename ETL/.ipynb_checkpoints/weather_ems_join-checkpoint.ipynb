```python
weather = pd.read_csv('../data/cntrlprk_weather.csv')
```

```python
weather.head()
```

```python
weather.drop(columns=['Unnamed: 0','STATION','NAME','DATE'], inplace=True)
```

```python
weather.info()
```

```python
ems_weather = pd.merge(ems, weather,  how='left', left_on=['CALL_MONTH','CALL_DAY','CALL_YEAR'], right_on = ['MONTH','DAY','YEAR'])
```

```python
ems_weather.head()
```

```python
ems_weather.drop(columns=['MONTH','DAY','YEAR'], inplace=True)
```

```python
ems_weather.head()
```

```python
ems_weather.to_csv('../data/ems_weather.csv')
```
